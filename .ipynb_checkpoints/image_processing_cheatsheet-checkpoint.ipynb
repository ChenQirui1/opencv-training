## Changing color space

```python
import numpy as np
import cv2 as cv

from matplotlib import pyplot as plt
```

```python
def bgr2rgb(img):
    return cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
#available color changes
flags = [i for i in dir(cv) if i.startswith('COLOR_')]
flags
```

```python
#preparing images
img = cv.imread('apple.jpg')
assert img is not None

grayscale_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

binary_j_img = cv.imread('alpha_j.png',cv.IMREAD_GRAYSCALE)

print("img: ", img.shape)
print("grayscale: ", grayscale_img.shape)
print("binary: ", binary_j_img.shape)
```

### Scaling

```python
#return same type as original img
res = cv.resize(img, None, fx=2,fy=2, interpolation = cv.INTER_CUBIC)

#prefered interpolation
res_shrink = cv.resize(img, None, fx=0.5,fy=0.5, interpolation = cv.INTER_AREA)

res_zoom_slow = cv.resize(img, None, fx=2,fy=2, interpolation = cv.INTER_CUBIC)

## used as the general one
res_zoom_faster = cv.resize(img, None, fx=2,fy=2, interpolation = cv.INTER_LINEAR)

print("Original img: ", img.shape)
print("Shrink img: ", res_shrink.shape)
print("Zoom Cubic img: ", res_zoom_slow.shape)
print("Zoom Linear img: ", res_zoom_faster.shape)
```

### Translation

```python
# matrix for translation

rows, cols = grayscale_img.shape

# M 
M = np.float32([[0,1,0],[1,0,0]])

dst = cv.warpAffine(img,M,(cols,rows))
```

```python
%matplotlib inline
plt.imshow(bgr2rgb(dst))
plt.show()
```

```python
dst.shape
```

```python

# i can also rotate the image with a n
rows, cols = grayscale_img.shape

# M 
M = np.float32([[0,1,0],[1,0,0]])

dst = cv.warpAffine(img,M,(cols,rows))
```

```python
# cols-1 and rows-1 are the coordinate limits.

# getRotation is get the matrix to transform, then apply to warpAffine for transforming
M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),40,1)
dst = cv.warpAffine(img,M,(cols,rows))
```

```python
%matplotlib inline
plt.imshow(bgr2rgb(dst))
plt.show()
```

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
 
M = cv.getAffineTransform(pts1,pts2)
 
dst = cv.warpAffine(img,M,(cols,rows))
 
plt.subplot(121),plt.imshow(bgr2rgb(img)),plt.title('Input')
plt.subplot(122),plt.imshow(bgr2rgb(dst)),plt.title('Output')
plt.show()
```

```python
rows,cols,ch = img.shape
 
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
 
M = cv.getPerspectiveTransform(pts1,pts2)
 
dst = cv.warpPerspective(img,M,(300,300))
 
plt.subplot(121),plt.imshow(bgr2rgb(img)),plt.title('Input')
plt.subplot(122),plt.imshow(bgr2rgb(dst)),plt.title('Output')
plt.show()
```

## Image Thresholding
- Simple
- Adaptive
- Otsu's Binarization

https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html

Only takes in greyscale, not useful to implement it for color images

```python
#simple thresholding
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

#img = cv.imread('gradient.png', cv.IMREAD_GRAYSCALE) #why imread_grayscale?


"""
cv.threshold
The first argument is the source image, which should be a grayscale image. 
The second argument is the threshold value which is used to classify the pixel values. 
The third argument is the maximum value which is assigned to pixel values exceeding the threshold
"""

# threshold types: https://docs.opencv.org/4.x/d7/d1b/group__imgproc__misc.html#gaa9e58d2860d4afa658ef70a9b1115576
ret,thresh1 = cv.threshold(grayscale_img,127,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(grayscale_img,127,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(grayscale_img,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(grayscale_img,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(grayscale_img,127,255,cv.THRESH_TOZERO_INV)
 
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [grayscale_img, thresh1, thresh2, thresh3, thresh4, thresh5]
 
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
 
plt.show()
```

```python
#adaptive thresholding

ret,th1 = cv.threshold(grayscale_img,127,255,cv.THRESH_BINARY)
th2 = cv.adaptiveThreshold(grayscale_img,255,cv.ADAPTIVE_THRESH_MEAN_C,\
            cv.THRESH_BINARY,11,2)
th3 = cv.adaptiveThreshold(grayscale_img,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv.THRESH_BINARY,11,2)
 
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [grayscale_img, th1, th2, th3]
 
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
# Otsu's Binarization
# Otsu's method avoids having to choose a value and determines it automatically.


# global thresholding
ret1,th1 = cv.threshold(grayscale_img,127,255,cv.THRESH_BINARY)
 
# Otsu's thresholding
ret2,th2 = cv.threshold(grayscale_img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
 
# Otsu's thresholding after Gaussian filtering
blur = cv.GaussianBlur(grayscale_img,(5,5),0)
ret3,th3 = cv.threshold(blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
 
# plot all the images and their histograms
images = [grayscale_img, 0, th1,
          grayscale_img, 0, th2,
          blur, 0, th3]
titles = ['Original Noisy Image','Histogram','Global Thresholding (v=127)',
          'Original Noisy Image','Histogram',"Otsu's Thresholding",
          'Gaussian filtered Image','Histogram',"Otsu's Thresholding"]
 
for i in range(3):
    plt.subplot(3,3,i*3+1),plt.imshow(images[i*3],'gray')
    plt.title(titles[i*3]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,3,i*3+2),plt.hist(images[i*3].ravel(),256)
    plt.title(titles[i*3+1]), plt.xticks([]), plt.yticks([])
    plt.subplot(3,3,i*3+3),plt.imshow(images[i*3+2],'gray')
    plt.title(titles[i*3+2]), plt.xticks([]), plt.yticks([])
plt.show()
```

## Image Smoothing

low-pass filters

https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html

- Averaging (Mean)
- Gaussian
- Median
- Bilateral

```python
# core
# the methods uses image filter to perform convolution

# e.g. 5x5 filter of ones 
kernel = np.ones((5,5),np.float32)/25

dst = cv.filter2D(img,-1,kernel)
```

```python
#methods

#averaging
avg_blur = cv.blur(img,(5,5))

#guassian, effective for removing guassian noise
#if want kernel, use cv.getGaussianKernel()
#can specify the sigma x/y if distribution known
guassian_blur = cv.GaussianBlur(img,(5,5),0)

#median, highly effective against salt-and-pepper noise
median_blur = cv.medianBlur(img,5)

#bilaternal, slow but highly effective, while keeping edges sharp
bilateral_blur = cv.bilateralFilter(img,9,75,75)

# plot all the images and their histograms
images = [avg_blur,guassian_blur,median_blur,bilateral_blur]
titles = ['Averaging','Gaussian','Median',
          'Bilateral']

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(bgr2rgb(images[i]))
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

## Morphological Transformations

simple operations based on the image shape. It is normally performed on **binary images**. It needs two inputs, one is our original image, second one is called structuring element or kernel which decides the nature of operation

https://docs.opencv.org/4.x/d9/d61/tutorial_py_morphological_ops.html

```python
#structuring kernel
kernel = np.ones((5,5),np.uint8)

kernel_99 = np.ones((9,9),np.uint8)

"""
in some cases, you may need elliptical/circular shaped kernels, use cv.getStructuringElement()
"""
```

```python
## methods

# erosion 
# erodes away the boundaries of foreground object (Always try to keep foreground in white)
# ! useful for removing small white noises
erosion = cv.erode(binary_j_img,kernel,iterations = 1)

# dilation -> opposite of erosion
"""
in cases like noise removal, erosion is followed by dilation
erosion removes white noises, but also shrinks our object. So we dilate it.

also useful in joining broken parts of an object
"""
dilation = cv.dilate(binary_j_img,kernel,iterations = 1)

###

# opening -> basically erosion followed by dilation
# use in removing noise
opening = cv.morphologyEx(binary_j_img, cv.MORPH_OPEN, kernel)

# closing -> dilation followed by Erosion
# useful in closing small holes inside the foreground objects, or small black points on the object
closing = cv.morphologyEx(binary_j_img, cv.MORPH_CLOSE, kernel)

###

# difference between dilation and erosion of an image
# outline of object
gradient = cv.morphologyEx(binary_j_img, cv.MORPH_GRADIENT, kernel)

### the following using 9x9 kernel

# top hat -> difference between input image and Opening of the image
tophat = cv.morphologyEx(binary_j_img, cv.MORPH_TOPHAT, kernel_99)

# black hat -> difference between the closing of the input image and input image
blackhat = cv.morphologyEx(binary_j_img, cv.MORPH_BLACKHAT, kernel_99)

###

# plot all the images
images = [binary_j_img, erosion, dilation,opening,closing, gradient, tophat, blackhat]
titles = ['Original','Erosion','Dilation','Opening',
          'Closing','Gradient','Tophat','Blackhat']

for i in range(8):
    plt.subplot(2,4,i+1),plt.imshow(bgr2rgb(images[i]))
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

## Histograms in OpenCV


https://docs.opencv.org/4.x/de/db2/tutorial_py_table_of_contents_histograms.html

More on image histograms: https://www.cambridgeincolour.com/tutorials/histograms1.htm


```python
#ploting histogram

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()

# the lines seem to overlap quite evenly, could be that is it mostly gray?
```

```python
#ploting 2d histogram

hsv = cv.cvtColor(img,cv.COLOR_BGR2HSV)
hist = cv.calcHist( [hsv], [0, 1], None, [180, 256], [0, 180, 0, 256] )
 
plt.imshow(hist,interpolation = 'nearest')
plt.show()

#based on this, there is 3 peak near S = 10, ( H= 30 , 75, 115)
```

### Histogram Equalization

- improve the contrast of our images by stretch the histogram


```python
# numpy implementation
hist,bins = np.histogram(grayscale_img.flatten(),256,[0,256])
 
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
 
plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
#histogram equalization
equ = cv.equalizeHist(grayscale_img)

#clahe
# create a CLAHE object (Arguments are optional).
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(grayscale_img)

# plot all the images and their histograms
images = [grayscale_img,equ,cl1]
titles = ['Original','Histogram Equalization','CLAHE']

for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(bgr2rgb(images[i]))
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
